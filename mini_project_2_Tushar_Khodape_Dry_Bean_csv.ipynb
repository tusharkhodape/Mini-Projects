```python
plt.figure(figsize=(10,8))
wedges,labels,pcts = plt.pie(x = data['Class'].value_counts().values,labels=data['Class'].value_counts().index, autopct="%.2f")
for l in labels:
    l.set_fontsize(16)
for p in pcts:
    p.set_fontsize(16)
plt.title("Distribution of class", fontsize = 16)
plt.show()
```

## Distribution of numerical Features

```python
Numeric_cols = data.drop(columns=['Class']).columns

fig, ax = plt.subplots(4, 4, figsize=(15, 12))
for variable, subplot in zip(Numeric_cols, ax.flatten()):
    g=sns.histplot(data[variable],bins=30, kde=True, ax=subplot)
    g.lines[0].set_color('crimson')
    g.axvline(x=data[variable].mean(), color='m', label='Mean', linestyle='--', linewidth=2)
plt.tight_layout()
```

### Insights:-
    1) Few distributions have long tails and most are bi-modal which means that some bean classes are different from others.
       

## Boxplot 

```python
fig, ax = plt.subplots(4, 4, figsize=(15, 12))

for variable, subplot in zip(Numeric_cols, ax.flatten()):
    sns.boxplot(y= data[variable], ax=subplot)
plt.tight_layout()
```

## Boxplot of numerical features for each type of bean:

```python
fig, ax = plt.subplots(8, 2, figsize=(15, 25))

for variable, subplot in zip(Numeric_cols, ax.flatten()):
    sns.boxplot(x=data['Class'], y= data[variable], ax=subplot)
plt.tight_layout()
```

### Observation:-
* We can see that the "Bombay" & "Horoz" classes are distinct from other classes.

# Classification techniques 

## Training and Testing Dataset

```python
!pip install sklearn
```

```python
x = data.drop(['Class'], axis=1)
y = data.Class
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

# Logistic Regression

```python
lm = LogisticRegression(random_state=20, n_jobs=5)
lm.fit(xtrain, ytrain)
```

## Prediction

```python
ypred_train = lm.predict(xtrain)
ypred_test = lm.predict(xtest)
```

## Confusion Matrix

```python
cm = confusion_matrix(ytest, ypred_test)
cm
```

```python
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlOrRd_r" ,fmt='g')
```

### Accuracy

```python
print("Accuracy of training data:", accuracy_score(ytrain, ypred_train)*100)
ac1 = accuracy_score(ytest, ypred_test)*100
print("Accuracy of testing data:", ac1)
```

### Classification Report


```python
target_names = ['Class']
print(classification_report(ytest, ypred_test))
```
